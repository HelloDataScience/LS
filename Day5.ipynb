```python
grp = mgd.groupby(by = '구')['거래금액'].mean().sort_values(ascending = False)
grp.head()
```

```python
pal = ['red'] * 3 + ['silver'] * 22
```

```python
# 95% 신뢰구간(confidential interval)
# 통계: 모집단 데이터의 특징(모수) <- 표본 통계량으로부터 모수를 추정
plt.figure(figsize = (8, 4))
sns.barplot(data = mgd, x = '구', y = '거래금액', color = '0.8', 
            estimator = 'mean', errorbar = None, order = grp.index, 
            hue = '구', palette = pal, hue_order = grp.index)
plt.xticks(rotation = 90, fontsize = 7)
plt.xlim(-1, 25)
plt.ylim(0, 22.5)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 0.1, s = f'{v:.1f}억', ha = 'center', va = 'bottom', 
             color = '0', fontsize = 7, fontweight = 'bold');
```

```python
scatter = dict(fc = '0.3', ec = '0.8', s = 20, alpha = 0.2)
regline = dict(color = 'blue', linewidth = 1.5, linestyle = '--')
```

```python
sns.regplot(data = gng, x = '전용면적', y = '거래금액', ci = None, 
            scatter_kws = scatter, line_kws = regline);
```

```python
cols = ['전용면적', '층', '세대수', '건축년도', '거래금액']
```

```python
sns.pairplot(data = gng[cols], 
             plot_kws = scatter, 
             diag_kws = dict(fc = '0.8'));
```

```python
sns.pairplot(data = gng, 
             x_vars = ['전용면적', '층', '세대수', '건축년도'], 
             y_vars = '거래금액', 
             kind = 'reg', 
             plot_kws = dict(scatter_kws = scatter, 
                             line_kws = regline), 
             diag_kws = dict(fc = '0.8'));
```
