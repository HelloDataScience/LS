```python
sns.histplot(data = mgd, x = '거래금액', binwidth = 2, binrange = (0, 60), 
             hue = '금액구간', hue_order = ['5천 미만', '5천 이상'], 
             ec = '0', palette = pal)
plt.title(label = '아파트 거래금액의 분포', size = 16, color = 'red')
plt.xlabel(xlabel = '거래금액(억)', size = 12, color = 'blue')
plt.ylabel(ylabel = '거래건수(건)', size = 12, color = 'blue');
```

```python
sns.histplot(data = mgd, x = '거래금액', binwidth = 2, binrange = (0, 60), 
             fc = '1', ec = '0.8', stat = 'density')
sns.kdeplot(data = mgd, x = '거래금액', c = 'red', lw = 0.5, ls = '--')
plt.xlim(-3, 63);
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

sns.histplot(data = mgd, x = '거래금액', binwidth = 2, binrange = (0, 60), 
             fc = '1', ec = '0.8', ax = ax1)
sns.kdeplot(data = mgd, x = '거래금액', c = 'red', lw = 0.5, ls = '--', ax = ax2)
plt.xlim(-3, 63);
```

```python
cond = mgd['구'].str.contains(pat = '강[남동북]')
sub = mgd.loc[cond, :]
```

```python
pd.pivot_table(data = sub, values = '거래금액', index = '구', 
               aggfunc = ['count', 'mean', 'std'])
```

```python
sns.histplot(data = sub, x = '거래금액', binwidth = 2, binrange = (0, 60), 
             hue = '구', ec = '0', palette = pal, 
             hue_order = ['강북구', '강동구', '강남구']);
```

```python
sns.kdeplot(data = sub, x = '거래금액', hue = '구', fill = True, 
            palette = pal, hue_order = ['강북구', '강동구', '강남구'])
plt.xlim(-3, 63);
```

```python
outlier = {
    'marker': 'o', 
    'markersize': 3, 
    'markerfacecolor': 'pink', 
    'markeredgecolor': 'red', 
    'markeredgewidth': 0.5
}
```

```python
sns.boxplot(data = mgd, y = '거래금액', color = '0.8', linewidth = 0.5, 
            flierprops = outlier);
```

```python
grp = mgd.groupby(by = '구')['거래금액'].median().sort_values()
grp.head()
```

```python
sns.boxplot(data = mgd, x = '구', y = '거래금액', color = '0.8', linewidth = 0.5, 
            flierprops = outlier, order = grp.index)
plt.xticks(rotation = 90, fontsize = 7);
```
